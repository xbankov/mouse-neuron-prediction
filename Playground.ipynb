```python
from google.colab import drive
drive.mount('/content/drive')

path = Path('drive/My Drive/Colab Notebooks/Neuroscience/')
```

```python
data = sio.loadmat(path/f'data/normal/natimg2800_M170717_MP034_2017-09-11.mat')
spatial = data['med']
resp = data['stim'][0]['resp'][0]
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

xs = spatial[:, 0]
ys = spatial[:, 1]
zs = spatial[:, 2]
c = resp[2940, :]

normalized = (c - np.mean(c))/ np.std(c)

ax.scatter(xs, ys, zs, c=normalized, marker='o', cmap='hot')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.savefig(path/'hot_map_invert.jpg')
```
